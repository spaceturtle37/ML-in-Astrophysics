A few weeks ago, you worked on finding periodic planet signals in the light curve of Kepler-90, a star that is photometrically stable. The periodogram worked nicely because 
1. we cleaned the light curve for you to squelch red noise, and

Life gets harder when the star itself has "quasi-periodic" variations. This means that the lightcurve of the star repeats itself, but not always with the same period! (Physically, this can happen when a star is rotating and has a significant magnetic field.)

In the cells below, we will load the lightcurve of the quasi-periodic star "KIC 2157356".

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from astropy.table import Table
tab = Table.read('KIC2157356.txt',format='ascii')
tab['quarter'] = tab['quarter'].astype('int')
tab
```

```python
plt.figure(figsize=(8,5)).patch.set_facecolor('w')

qs = sorted(np.unique(tab['quarter']))
means = []
cycle_map = {}
for i, q in enumerate(qs):
    ind = tab['quarter']==q
    t = tab[ind]
    plt.errorbar(t['time'],t['flux'], yerr=t['error'], marker='.', linestyle='None', alpha=0.01)
    meanflux = np.mean(t['flux'])
    cycle_map[q] = ind
    means.append(meanflux)
    if i == 0:
        plt.axhline(meanflux, label='Camera 0: Mean = $m$', color='grey', ls=":")
    else:

        vmin = means[0]
        vmax = meanflux

        plt.plot((t['time'][0], t['time'][0]), (vmin, vmax), label=rf'Camera {i}: Mean = $m+c_{i}$', color=f'C{i}', ls='--') 
    
plt.xlabel('Time')
plt.ylabel('Flux')
plt.legend(frameon=False)
plt.title('Lightcurve of KIC 2157356 in 4 different observation cycles')
```

In the cell above, we have plotted Kepler data on the same star (KIC 2157356) over four different observation periods. There are two things you should note about the plot:

1. The star exhibits some periodic behavior, but it is not perfectly regular (like you've seen in other problems).
2. Kepler has observed this star with four different cameras, or in four different "observation cycles". In each camera, the average flux measured is different! This isn't because the star's flux is changing. Instead, Kepler's cameras are just very difficult to calibrate correctly. Each adds a different constant offset to the observed flux of KIC 2157356: In the plot above, camera 1's measured flux is offset from camera 0's by $c_1$, camera 2's from camera 0's by $c_2$, and camera 3's from camera 0's by $c_3$.

You'll need **four model parameters** ($m, c_1, c_2, c_3$) to model the mean flux of the star in camera 0 ($m$), as well as offsets from this mean in other cameras ($c_1, c_2, c_3$).

To model the periodic behavior of the lightcurve, it will not be enough to use methods from previous homeworks -- those methods only work when periodic behavior is very regular, and this star's behavior is not! Instead we'll model this light curve as a **Gaussian process.** Recall that Gaussian processes require a covariance function (or kernel) that describes the relationships between fluxes measured at different times.

For this problem, we'll model the covariance using "something periodic" and "something that varies the periodicity smoothly": The kernel $k(t_i, t_j)$ will be

### $$k(t_i, t_j) = A\cdot \exp\left(-\Gamma_1\cdot \sin^2\left(\frac{\pi}{P}|t_i - t_j|\right)\right) \cdot \exp\left(-\frac{|t_i-t_j|^2}{\lambda}) \right) $$

(A quick overview: The first $\exp$ term is called an "exponential sine squared" kernel, and will model the periodic components of the model. The second $\exp$ term is an "exponential squared" kernel and it will model the smooth variation.)

Between the mean flux in camera 0, the offsets in the other cameras, and the kernel function, **your model will have 8 parameters to fit:**
### $$m, c_1, c_2, c_3, A, \Gamma_1, P, \lambda$$

### Q1: To implement the GP correlations, use the `george` package to construct this quasi-periodic kernel

The kernels in the `george` package are documented here: https://george.readthedocs.io/en/latest/user/kernels/. 

In particular, you should combine the kernels `ExpSine2Kernel` and `ExpSquaredKernel` to generate the kernel that was presented in the introduction.

```python
import george
from george import kernels

def kernel(A, gamma,logP, lam):
    return A*kernels.ExpSine2Kernel(gamma=gamma, log_period=logP)*kernels.ExpSquaredKernel(metric=lam)

def kernel2(A, gamma,logP, lam):
    return kernels.ExpSquaredKernel(metric=lam)
```

### To define the cycles, we'll figure out the start time of each cycle, which really defines the end time of the constant value for the previous cycle

```python
transition_times = {key: tab[ind]['time'].min() for key, ind in cycle_map.items()}
print(transition_times)
```

```python
from george.modeling import Model
class kepler_mean_model(Model):
    parameter_names = ("m", "c1", "c2", "c3")
    def get_value(self, t):
        
        t = t.flatten()
        # the baseline model is the mean m
        mean_model = np.repeat(self.m, len(t))
        
        # each cycle after 6 has it's own offset
        # you could do this with a single dictionary/list comprehension 
        ind = (t >= transition_times[7]) & (t < transition_times[8])
        mean_model[ind] += self.c1
        ind = (t >= transition_times[8]) & (t < transition_times[9])
        mean_model[ind] += self.c2
        ind = t >= transition_times[9]
        mean_model[ind] += self.c3
        
        return mean_model       
```

### Q3: With your model and likelihood constructured, write down priors on the parameters (you should be able to estimate from the plots)

```python
# First we'll do initial guesses:

# means and offsets are easy
m = means[0]
c1, c2, c3 = means[1:] - m

# just use the stddev for the first cycle to get some scale for the amplitude
A = tab['flux'][cycle_map[6]].std() 

# gamma is modulating the periodic variation - we'll just set it to 1.
gamma= 1.

# squiggles are roughly every 15 days from the plot
logP = np.log(15.)

# long-period variations are on the scale of the Kepler cycles - from cycle 8/9, this is maybe 150 days
lam  = 150

p0 = [m, c1, c2, c3, A, gamma, logP, lam]
```

```python
everyn = 30
t = tab['time'][::everyn]
y = tab['flux'][::everyn]
yerr = tab['error'][::everyn]

model = george.GP(mean=kepler_mean_model(m, c1, c2, c3), kernel=kernel2(A, gamma, logP, lam))
model.compute(t)
pred1 = model.predict(y, tab['time'], return_cov=False)

model.compute(t, yerr)
pred2 = model.predict(y, tab['time'], return_cov=False)


#plt.plot(tab['time'], tab['flux'], 'r.')
plt.errorbar(tab['time'], pred1-pred2, yerr=tab['error'],color='k', ms=0.5, alpha=0.5)
#plt.errorbar(tab['time'], pred1, yerr=tab['error'],color='k', ms=0.5, alpha=0.5)
#plt.errorbar(tab['time'], pred2, yerr=tab['error'],color='k', ms=0.5, alpha=0.5)
#plt.ylim(5000,6000)
#plt.ylim(5000,6000)
ymax = 100
plt.ylim(-ymax,ymax)
```

### OK, so even with initial guesses and using only every 30th observation, this is reasonable looking. On to actual priors - we'll define bounds and Gaussian sigmas - note that some of the parameters are in log

```python
guess = [m, c1, c2, c3, np.log(A), gamma, logP, np.log(lam)]

# we can do a tophat on each parameter minimally
bounds = []

# we already have a scale of variation for the mean and offsets
bounds.append((m - 2*A, m+2*A))
bounds.append((c1 - 2*A, c1+2*A))
bounds.append((c2 - 2*A, c2+2*A))
bounds.append((c3 - 2*A, c3+2*A))
# we have some additional constraints, but we'll define this in the log prior function

# don't allow the amplitude to go to 0 - the scale can't be too far off from our initial guess
bounds.append((np.log(0.5*A), np.log(1000*A)))

# no good idea for gamma, but probably shouldn't allow it to be 0
bounds.append((0.1, 100.))

# logP - from the plot, 15 is pretty reasonable - it's not 10 and it's not 20, so lets use those as bounds
bounds.append((np.log(10.), np.log(20.)))

# lam - we know that this is longer than the periodic variations 
bounds.append((np.log(20.), np.log(1000.)))



# we also need to initialize the walkers for emcee so we need some scale parameter
sigma = []

# mean and offset is again easy
sigma.append(10.)
sigma.append(10.)
sigma.append(10.)
sigma.append(10.)

# A
sigma.append(np.log(5.))

# gamma
sigma.append(0.1)

# logP
sigma.append(np.log(2.))

# lambda
sigma.append(np.log(20.))
```

```python
# so finally our prior is:
def log_prior(p):
    global bounds
    logprior = np.zeros(8)
    for j in range(8):
        limit = bounds[j]
        if (p[j] < limit[0]) or (p[j] > limit[1]):
            return -np.inf
        else:
            logprior[j] = -0.5*((p[j]-guess[j])/(5.*sigma[j]))**2
    
    # force the ordering of the offsets between the cycles
    # this is mostly relevant for cycle 9 - the others should be imposed by the bounds
    if p[1] > p[2]:
        return -np.inf
    
    if p[2] < p[3]:
        return -np.inf
    
    if p[1] > p[3]:
        return -np.inf
        
    return logprior.sum()
```

### Q4: Use `emcee` to optimize the model parameters and hyper-parameters, **using only every 10th sample in time**
### (Don't go overboard with the number of walkers or steps)
https://george.readthedocs.io/en/latest/tutorials/hyper/ may help

```python
import scipy.stats as st
import emcee
```

```python
# setup the posterior
def lnprob(p):
    model.set_parameter_vector(p)
    return model.log_likelihood(y, quiet=True) + log_prior(p)

# setup the sampler
nwalkers, ndim = 32, len(model)
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob)

# initalize walker positions
p0 = st.norm.rvs(loc=guess, scale=sigma, size=(nwalkers,ndim))

# number of samples
nburn = 500
nsamp = 1000
_ = sampler.run_mcmc(p0, nburn+nsamp, progress=True)
```

### Q5: Plot your posterior model over the data after correcting for the offsets, showing the points you used to condition the GP in red, and the remaining data in black. 

```python
thin = 10
sample = sampler.chain[:, nburn:, :].reshape(-1, ndim)
sample = sample[::thin,:]
```

```python
import corner
labels = ['m','c1','c2','c3','logA','gamma','logP',r'log$\lambda$']

fig = corner.corner(sample,labels=labels, plot_contours=True, show_titles=True, smooth=1)
fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,figsize=(14,5))
tgrid = np.linspace(450, 1000, 1001)
stats = np.percentile(sample, [50, 16, 84], axis=0)
res = stats[0]
model.set_parameter_vector(res)

yfull, cov = model.predict(y, tgrid, return_cov=True)
err = np.sqrt(np.diag(cov))

tall = tab['time']
yres = y - model.mean.get_value(t)
yall = tab['flux']-model.mean.get_value(tall)
ypred = yfull-model.mean.get_value(tgrid)


plt.plot(t, yres, 'ro', label='points used to condition GP')
plt.plot(tall, yall, 'k.', ms=3, alpha=0.5, label='All Data')
plt.plot(tgrid, ypred, color='C0',linestyle='--', label='Median of GP posterior')
plt.fill_between(tgrid, ypred-err, ypred+err, color='grey', alpha=0.3)

ax.set_xlabel('MJD')
ax.set_ylabel('Flux - Offsets')
ax.legend(frameon=False)
    
fig.tight_layout()
```
